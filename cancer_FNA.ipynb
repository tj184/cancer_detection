```python
import pandas as pd 
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
data=pd.read_csv("C:\\Users\\Lenovo\\Downloads\\data.csv")
data
```

```python
data=data.drop(['Unnamed: 32'],axis=1)
data=data.drop(['id'],axis=1)
data
```

```python
y=data['diagnosis']
X=data.drop(['diagnosis'],axis=1)
X.shape,y.shape
```

```python
for col in X.columns:
        max_value = X[col].max()
        X[col] = X[col] / max_value
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder
y = np.array(y).reshape(-1, 1)


encoder = OneHotEncoder(sparse_output=False)


y_en = encoder.fit_transform(y)
```

```python
print(y_en)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_en, test_size=0.25, random_state=42)
X_train.shape,y_train.shape
```

```python
import os
model = keras.Sequential([
        keras.layers.Dense(20,input_shape=(30, ), activation='tanh'),
        keras.layers.Dense(2, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
log_dir = "logs/"  
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(X_train, y_train, epochs=200,callbacks=[tensorboard_callback])
       
```

```python
model.predict(X_test)
```
